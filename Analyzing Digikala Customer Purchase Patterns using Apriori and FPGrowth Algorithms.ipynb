```python
# Required libraries for Apriori amd FPGrowth algorithms

import efficient_apriori
from mlxtend.frequent_patterns import apriori, fpgrowth
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder

# Importing pandas library for doing process on dataframes
import pandas as pd
```

```python
df = pd.read_csv('orders.csv')
df
```

# Organizing the dataset with ID_Customer and DataTime of Purchase

```python
df_group = df.groupby(['ID_Customer','DateTime_CartFinalize'])['ID_Item'].apply(list)
df_group
```

# Creating list of items 

```python
l = [[str(i) for i in x] for x in df_group.values]
```

# Applying Encoder class for transaction data in Python lists

```python
TE = TransactionEncoder()
TE_Fit = TE.fit(l).transform(l)
dataframe = pd.DataFrame(TE_Fit, columns=TE.columns_)
dataframe
```

# For doing easier process , we select limited number of our data and avoid using much memory

```python
TE = TransactionEncoder()
TE_Fit = TE.fit(l[:1000]).transform(l[:1000])
dataframe = pd.DataFrame(TE_Fit, columns=TE.columns_)
dataframe
```

# Finding frequent_itemset_apriori from apriori libraries

```python
frequent_itemsets_apriori = apriori(dataframe, min_support=0.001, use_colnames=True)
frequent_itemsets_apriori
```

# Define a function for calculating X2, Imbalance Ratio and Lift 

```python
def x2_IR(df):
    dataframe['X2'] = dataframe['leverage'] ** 2 / df['support']
    dataframe['IR'] = abs(df['antecedent support'] - df['consequent support']) / (df['antecedent support'] + df['consequent support'] - df['support'])
    return df[[ 'consequents','antecedents', 'confidence', 'support','lift', 'X2', 'IR']]
```

# Showing the result

```python
result = association_rules(frequent_itemsets_apriori, metric='lift', min_threshold=0.01)
result = x2_IR(result)
result
```

# Showing Significant Patterns

```python
def significant(df):

    #IR
    print('\nrules with non-zero imbalance ratio (IR):\n')
    print(df[df['IR'] != 0])

    #Lift
    print('\nrules with maximum lift:\n')
    print(df[df['lift'] == df['lift'].max()])

    # X^2
    print('\nrules with maximum x2 (correlation):\n')
    print(df[df['X2'] == df['X2'].max()])
```

```python
significant(result)
```

# Define function for proccessing Tehran city dataset

```python
def process(df, city, n=0):
    
    df1 = df[df['city_name_fa'] == city]
    df_g = df1.groupby(['ID_Customer','DateTime_CartFinalize'])['ID_Item'].apply(list)

    l = [[str(i) for i in x] for x in df_g.values]
    print(f'number of items: {len(l)}')
    
    te = TransactionEncoder()
    if n:
        te_ary = te.fit(l[:n]).transform(l[:n])
    else:
        te_ary = te.fit(l).transform(l)
    df_out = pd.DataFrame(te_ary, columns=te.columns_)
    return df_out
```
